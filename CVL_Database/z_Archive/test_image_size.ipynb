```python
from PIL import Image, ImageOps
import os
import random
import matplotlib.pyplot as plt
```

```python
def read_resize_and_display_image(path, width, height, images_to_display):
    #Get random folder
    random_folder = random.sample(os.listdir(path), 1)
    
    #How many images in folders
    img_count = os.listdir(path +  "/" + random_folder[0])
    
    #Select random images
    images = random.sample(img_count, images_to_display)
    
    #open image, convert to grayscale, resize and plot it
    plt.figure(figsize=(12,7))
    for i in range(images_to_display):
        image = Image.open((path + "/" + random_folder[0] + "/" + images[i]))
        gray = ImageOps.grayscale(image)
        resized_image = gray.resize((width, height), Image.ANTIALIAS)
        ax = plt.subplot(len(images),1,  i+1)
        resized_image.save(f"image{i}", "JPEG")
        plt.imshow(resized_image, cmap="gray")
```

```python
train_path = "trainset/lines"
width = 150
height = 150
images_to_display = 2
read_resize_and_display_image(train_path, width, height, 5)
```
