```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
train_path = "trainset/lines"
test_path = "testset/lines"
```

```python
def view_random_image_pair(path, images_to_display):
    #Get random folder
    random_folder = random.sample(os.listdir(path), 1)
    
    #How many images in folders
    img_count = os.listdir(path +  "/" + random_folder[0])
    
    #Select random images
    images = random.sample(img_count, images_to_display)
    
    plt.figure(figsize=(12,7))
    for i in range(len(images)):
        #load and display the image
        ax = plt.subplot(len(images),1,  i+1)
        img = mpimg.imread(path + "/" + random_folder[0] + "/" + images[i])
        plt.title(path + "/" + random_folder[0] + "/" + images[i])
        plt.imshow(img)
```

```python
#Display lines in train path
view_random_image_pair(train_path, 5)
```

```python
#Display lines in train path
view_random_image_pair(test_path, 5)
```
