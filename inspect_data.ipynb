```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from paths import dataset_paths
```

```python
def view_random_image_pair(path, images_to_display):
    #select random folder
    folder = random.choice(path)
    #select random images
    images = random.sample(os.listdir(folder), images_to_display)
    
    plt.figure(figsize=(12, 7))
    for i in range(len(images)):
        ax = plt.subplot(len(images),1,  i+1)
        img = mpimg.imread(folder + "/" + images[i])
        plt.title(images[i])
        plt.imshow(img)
```

```python
view_random_image_pair(dataset_paths, 5)
```
