```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
path_Data_A = "Dataset_22092021_lina/SCID/Data_A/processed/rgb_rows"
path_Data_B = "Dataset_22092021_lina/SCID/Data_B/processed/rgb_rows"
```

```python
def view_random_image_pair(path, images_to_display):
    #Get random ccl folder
    random_folder = random.sample(os.listdir(path), 1)
    
    #List folders in parent folder
    subfolder = os.listdir(path +  "/" + random_folder[0] + "/" + "rows")
    
    #Get only folders with labeled hands
    hands_list = []
    for i in range(len(subfolder)):
        if subfolder[i].startswith("hand"):
            hands_list.append(subfolder[i])
        else:
            pass
        
    #get a random folder for hands
    hands_folder = random.choice(hands_list)

    print(hands_folder)
    #how many images in hands folder
    img_count = os.listdir(path + "/" + random_folder[0] + "/" + "rows" + "/" + hands_folder)

    #Select random images
    images = random.sample(img_count, images_to_display)
    
    plt.figure(figsize=(12, 7))
    for i in range(len(images)):
        #load and display the images
        ax = plt.subplot(len(images),1,  i+1)
        img = mpimg.imread(path + "/" + random_folder[0] + "/" + "rows" + "/" + hands_folder + "/" + images[i])
        plt.title(path + "/" + random_folder[0] + "/" + "rows" + "/" + hands_folder + "/" + images[i])
        plt.imshow(img)
```

```python
view_random_image_pair(path_Data_A, 5)
```

```python
view_random_image_pair(path_Data_B, 5)
```
